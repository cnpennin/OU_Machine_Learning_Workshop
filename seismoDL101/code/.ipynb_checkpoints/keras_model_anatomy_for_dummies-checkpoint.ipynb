```python
import numpy as np
import pylab as plt
import warnings
import os
import h5py
import matplotlib as mpl
%matplotlib inline

import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv1D, MaxPooling1D, LSTM
from keras import losses
from keras.utils.np_utils import to_categorical
from keras.layers.normalization import BatchNormalization
```

```python
model3 = Sequential()
    
nsamp = 400
ncomp = 3

#keras.layers.Conv1D(filters, kernel_size, strides=1)
model3.add(Conv1D(4, 8, padding='same', input_shape=(nsamp, ncomp)))
model3.add(BatchNormalization())
model3.add(Activation('relu'))
model3.add(MaxPooling1D(pool_size=2))

model3.add(Flatten())  # write the output into a single 1D vector

model3.add(Dense(8))
model3.add(BatchNormalization())
model3.add(Activation('relu'))

model3.add(Dense(2))
model3.add(Activation('softmax'))

print(model3.summary())
```

### Address specific layers and look at content
##### Conv Layer

```python
# Convolutional layer
lconv = model3.layers[0]
weights = lconv.get_weights() # list of numpy arrays
nel = len(weights)
print('*weights* is a python-list of '+str(nel)+' np arrays')

w0  = weights[0]
ncoef, nin, nfilt = w0.shape
print('The first one has shape '+str(nw0))
print(' \ti.e. filters with '+str(ncoef)+' coefficients each',
      ' \n\tone for each of '+str(nin)+' input channels',
      ' \n\tand '+str(nfilt)+' filter channels (printed below)\n')


w1  = weights[1]
nw1 = w1.shape
print('The second one has shape '+str(nw1))
print(nw1)
print(w1)

print('')
print(w0)
```

##### Dense Layer

```python
lfc = model3.layers[5]
weights = lfc.get_weights() # list of numpy arrays
nel = len(weights)
print('*weights* is a python-list of '+str(nel)+' np arrays')

w0  = weights[0]
nw0 = w0.shape
print('The first one has shape '+str(nw0))

print('i.e. it is not nsamp*ncomp*nfilt / 2 (maxpool) = '+str(nsamp*ncomp*nfilt/2))
print('>> are the input channels combined before the dense layers?')
```
